```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Draw

all = pd.read_csv("/home/daan/tiberius_data/drugex3/ccr2/new_molecules/selected_molecules_fep.tsv", sep="\t")
#print(all.head())

unique = all.drop_duplicates(subset=["SMILES"])
print(unique.shape[0])

print(unique.head())

mols = [Chem.MolFromSmiles(x) for x in unique['SMILES']]
mols

#unique.to_csv("unique_1C.tsv", sep="\t")
```

```python
smoll = []
for mol in mols:
    if mol.GetNumAtoms() <= 28: 
        smoll.append(mol)
    #print(mol.GetNumAtoms())
    #molecule = (mol.GetNumAtoms() <= 27)
    #smoll.append(molecule)
smoll[0]
print(len(smoll))

img = Draw.MolsToGridImage(smoll, molsPerRow=3,  returnPNG=True, maxMols=156, useSVG=True)
img
```
